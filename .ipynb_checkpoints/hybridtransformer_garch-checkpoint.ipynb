```python
import numpy as np
import pandas as pd
import yfinance as yf

import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

from arch import arch_model

import torch
import torch.nn as nn
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

import shap
```

```python
sp500 = yf.download("^GSPC",
                    start="2015-01-01",
                    end="2025-12-31")

btc = yf.download("BTC-USD",
                  start="2015-01-01",
                  end="2025-12-31")

gold = yf.download("GC=F",
                   start="2015-01-01",
                   end="2025-12-31")

eurusd = yf.download("EURUSD=X",
                     start="2015-01-01",
                     end="2025-12-31")
```

```python
def compute_returns(df):

    returns = np.log(
        df['Close'] /
        df['Close'].shift(1)
    )

    return returns.dropna()

sp500_ret = compute_returns(sp500)
btc_ret = compute_returns(btc)
gold_ret = compute_returns(gold)
eur_ret = compute_returns(eurusd)
```

```python
garch = arch_model(
    sp500_ret.iloc[:, 0],
    vol='GARCH',
    p=1,
    q=1
)

result = garch.fit()

```

```python
volatility = result.conditional_volatility
plt.figure(figsize=(12,5))
plt.plot(volatility)
plt.title("GARCH Volatility")
plt.show()
plt.savefig("garch_volatility.png")
```

```python
features = pd.DataFrame()

features['returns'] = sp500_ret
features['volatility'] = volatility

features = features.dropna()
```

```python
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(features)
```

```python
def create_sequences(data,
                     seq_len=30):

    X = []
    y = []

    for i in range(len(data)-seq_len):

        X.append(
            data[i:i+seq_len]
        )

        y.append(
            data[i+seq_len,1]
        )

    return np.array(X), np.array(y)

X,y = create_sequences(
    scaled_data,
    30
)

```

```python
split = int(len(X)*0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
class TransformerModel(nn.Module):

    def __init__(
        self,
        input_dim=2,
        d_model=64,
        nhead=4,
        num_layers=2
    ):

        super().__init__()

        self.embedding = nn.Linear(
            input_dim,
            d_model
        )

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model,
            nhead=nhead,
            batch_first=True
        )

        self.transformer = nn.TransformerEncoder(
            encoder_layer,
            num_layers=num_layers
        )

        self.fc = nn.Linear(
            d_model,
            1
        )

    def forward(self,x):

        x = self.embedding(x)

        x = self.transformer(x)

        x = x[:,-1,:]

        return self.fc(x)
```

```python
model = TransformerModel()

criterion = nn.MSELoss()

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)
```

```python
for epoch in range(5):

    optimizer.zero_grad()

    preds = model(
        torch.FloatTensor(X_train)
    )

    loss = criterion(
        preds.squeeze(),
        torch.FloatTensor(y_train)
    )

    loss.backward()

    optimizer.step()

    print(epoch, loss.item())
```

```python
class MCDropoutModel(
    TransformerModel
):

    def __init__(self):

        super().__init__()

        self.dropout = nn.Dropout(
            0.3
        )
```

```python
predictions = []
input_tensor = torch.FloatTensor(X_test)

model.train()
with torch.no_grad():
    for _ in range(50):
        pred = model(input_tensor).squeeze().detach().cpu().numpy()
        predictions.append(pred)

predictions = np.array(predictions)

mean_pred = predictions.mean(axis=0)
std_pred = predictions.std(axis=0)

mean_pred = predictions.mean(axis=0)

std_pred = predictions.std(axis=0)
```

```python
lower = mean_pred - 1.96*std_pred
upper = mean_pred + 1.96*std_pred
```

```python
VaR95 = np.percentile(
    sp500_ret,
    5
)
print("VaR 95%:", VaR95)
```

```python
CVaR95 = sp500_ret[
    sp500_ret <= VaR95
].mean()

print("CVaR 95%:", CVaR95)
```

```python
explainer = shap.Explainer(
    model,
    torch.FloatTensor(X_train[:100])
)
shap_values = explainer(
    torch.FloatTensor(X_test[:50])
)

shap.summary_plot(
    shap_values
)
```

```python
mse = mean_squared_error(
    y_test,
    mean_pred
)
rmse = np.sqrt(mse)

mae = mean_absolute_error(
    y_test,
    mean_pred
)
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
```
